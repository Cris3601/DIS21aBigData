```python
# Imports

import os, shutil
import numpy as np

import matplotlib.pyplot as plt

import tensorflow
from tensorflow.keras import models, layers, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16

tensorflow.keras.__version__
```

```python
sample_number_train = 1000
sample_number_val = 250
sample_number_test = 243



# create directory structure
train_dir_original = "../data/original_train"
train_dir_benign_original = "../data/original_train/benign"  
train_dir_malignant_original = "../data/original_train/malignant"

train_dir = "../data/train"
train_dir_benign = "../data/train/benign"  
train_dir_malignant = "../data/train/malignant"  

validation_dir = "../data/validation"
validation_dir_benign = "../data/validation/benign"  
validation_dir_malignant = "../data/validation/malignant"

test_dir = "../data/test"
test_dir_benign = "../data/test/benign"  
test_dir_malignant = "../data/test/malignant" 



###### copy data from original_train to validation ######
### benign

# delete old images
filelist = [ f for f in os.listdir(train_dir_benign)]
for f in filelist:
    os.remove(os.path.join(train_dir_benign, f))

# create new images
fnames = ['{}.jpg'.format(i) for i in range(sample_number_train)] 
for fname in fnames:
    
    src = os.path.join(train_dir_benign_original, fname)
    dst = os.path.join(train_dir_benign, fname)
    
    if os.path.exists(src):
        shutil.copyfile(src, dst)


### malignant            
# delete old images
filelist = [ f for f in os.listdir(train_dir_malignant)]
for f in filelist:
    os.remove(os.path.join(train_dir_malignant, f))

# create new images
fnames = ['{}.jpg'.format(i) for i in range(sample_number_train)] 
for fname in fnames:
    
    src = os.path.join(train_dir_malignant_original, fname)
    dst = os.path.join(train_dir_malignant, fname)
    
    if os.path.exists(src):
        shutil.copyfile(src, dst)

        



 #### copy data from original_train to validation ######

### benign
# delete old images
filelist = [ f for f in os.listdir(validation_dir_benign)]
for f in filelist:
    os.remove(os.path.join(validation_dir_benign, f))
    
# create new images
fnames = ['{}.jpg'.format(i) for i in range(sample_number_train, sample_number_train+sample_number_val)]
for fname in fnames:
    
    src = os.path.join(train_dir_benign_original, fname)
    dst = os.path.join(validation_dir_benign, fname)
    
    if os.path.exists(src):
        shutil.copyfile(src, dst)
        


    

### malignant
# delete old images
filelist = [ f for f in os.listdir(validation_dir_malignant)]
for f in filelist:
    os.remove(os.path.join(validation_dir_malignant, f))
    
# create new images
fnames = ['{}.jpg'.format(i) for i in range(sample_number_train, sample_number_train+sample_number_val)]
for fname in fnames:
    src = os.path.join(train_dir_malignant_original, fname)
    dst = os.path.join(validation_dir_malignant, fname)
    
    if os.path.exists(src):
        shutil.copyfile(src, dst)
        
            
#### copy data from original_train to test ######

### benign
# delete old images
filelist = [ f for f in os.listdir(test_dir_benign)]
for f in filelist:
    os.remove(os.path.join(test_dir_benign, f))
    
# create new images
fnames = ['{}.jpg'.format(i) for i in range(sample_number_train+sample_number_val, sample_number_train+sample_number_val+sample_number_test)]
for fname in fnames:
    
    src = os.path.join(train_dir_benign_original, fname)
    dst = os.path.join(test_dir_benign, fname)
    
    if os.path.exists(src):
        shutil.copyfile(src, dst)
        
        


    

### malignant
# delete old images
filelist = [ f for f in os.listdir(test_dir_malignant)]
for f in filelist:
    os.remove(os.path.join(test_dir_malignant, f))
    
# create new images
fnames = ['{}.jpg'.format(i) for i in range(sample_number_train+sample_number_val, sample_number_train+sample_number_val+sample_number_test)]
for fname in fnames:
    src = os.path.join(train_dir_malignant_original, fname)
    dst = os.path.join(test_dir_malignant, fname)
    
    if os.path.exists(src):
        shutil.copyfile(src, dst)
        
```

```python
# print out current sample number of each directory
print('total original training benign images:', len(os.listdir(train_dir_benign_original)))
print('total original training malignant images:', len(os.listdir(train_dir_malignant_original)))
print('--------------------------')
print('total training benign images:', len(os.listdir(train_dir_benign)))
print('total training malignant images:', len(os.listdir(train_dir_malignant)))
print('--------------------------')
print('total validation benign images:', len(os.listdir(validation_dir_benign)))
print('total validation malignant images:', len(os.listdir(validation_dir_malignant)))
print('--------------------------')
print('total testing benign images:', len(os.listdir(test_dir_benign)))
print('total testing malignant images:', len(os.listdir(test_dir_malignant)))
```

```python
from tensorflow.keras.applications import VGG16
conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

# Methode B - 30 epochen

```python
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.summary()

conv_base.trainable = True # should be the default setting

print('This is the number of trainable weight tensors '
      '\033[4mbefore\033[0m freezing the conv_base:', len(model.trainable_weights))

# now freeze it
conv_base.trainable = False

print('This is the number of trainable weight tensors '
      '\033[4mafter\033[0m freezing the conv_base:', len(model.trainable_weights))

batch_z = 20

train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

# Note again, that the validation data should not be augmented!
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=batch_z,
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(224, 224),
        batch_size=batch_z,
        class_mode='binary')

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['acc'])

history = model.fit(
      train_generator,
      steps_per_epoch=30,
      epochs=30,
      validation_data=validation_generator,
      validation_steps=25,
      verbose=1)

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()

model.save('BenignMalignantClass_pretrained_CNN_FeatExtraction_using_data_augmentation_epochs_30_try_4.h5')
# del model
import pickle
class History_trained_model(object):
    def __init__(self, history, epoch, params):
        self.history = history
        self.epoch = epoch
        self.params = params

with open('BenignMalignantClass_pretrained_CNN_FeatExtraction_using_data_augmentation_epochs_30_try_4_History', 'wb') as file:
    model_history= History_trained_model(history.history, history.epoch, history.params)
    pickle.dump(model_history, file, pickle.HIGHEST_PROTOCOL)  
```
